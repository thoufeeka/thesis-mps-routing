```python
import pandas as pd
from qiskit import QuantumCircuit
from pathlib import Path

rows = []

for file in Path("../benchmarks").glob("*.qasm"):

    qc = QuantumCircuit.from_qasm_file(str(file))

    rows.append({
        "circuit": file.stem,
        "qubits": qc.num_qubits,
        "depth": qc.depth(),
        "gates": len(qc.data),
        "two_qubit_gates": qc.num_nonlocal_gates()
    })

df = pd.DataFrame(rows)

display(df)

df.to_csv(
    "../results/circuit_stats.csv",
    index=False
)
```
