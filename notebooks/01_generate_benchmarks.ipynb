```python
from mqt.bench import get_benchmark, BenchmarkLevel
from qiskit import qasm2
import os
```

```python
os.makedirs("../benchmarks", exist_ok=True)
```

```python
import inspect
print(inspect.signature(get_benchmark))
    
```

```python
for n in [10,20]:

    qc = get_benchmark(
        benchmark="qaoa",
        circuit_size=n,
        level=BenchmarkLevel.ALG
    )
    qc = qc.decompose(reps=20)
    print(qc.count_ops())
    with open(f"../benchmarks/qaoa_{n}.qasm","w") as f:
        f.write(qasm2.dumps(qc))
```

```python
for n in [10,20]:

    qc = get_benchmark(
        benchmark="qft",
        circuit_size=n,
        level=BenchmarkLevel.ALG
    )
    qc = qc.decompose(reps=10)
    with open(f"../benchmarks/qft_{n}.qasm","w") as f:
        f.write(qasm2.dumps(qc))
```

```python
vqe_variants = ["vqe_real_amp", "vqe_two_local"]

for variant in vqe_variants:
    for n in [10, 20]:
        print(f"Generating {variant} n={n}...")
        qc = get_benchmark(
            benchmark=variant,
            level=BenchmarkLevel.ALG,
            circuit_size=n,
        )
        print(f"  Circuit: {qc.num_qubits} qubits, {qc.depth()} depth")
        
        path = f"../benchmarks/vqe/{variant}_{n}.qasm"
        qasm_str = qasm2.dumps(qc)
        
        with open(path, "w") as f:
            f.write(qasm_str)
        
        print(f"  Written to: {path}")
```
